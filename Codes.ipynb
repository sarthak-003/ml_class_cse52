<h1>Gradient Descent</h1>

```python
import numpy as np
```

```python
data = np.array([
    [1, 2 ,4.029],
    [2, 3 ,6.9],
    [3, 6 ,12.401],
    [4, 8 ,15.873],
    [5, 10 ,20.23],
    [6, 12 ,24.019],
    [7, 14 ,27.691],
    [8, 16 ,33],
    [9, 18 ,36.21],
    [10, 20 ,41.074]
]) 
x = data[:, 0:2].reshape(2,-1)
y = data[:, 2] 
# y = 2*x1 + x2 + noise
d = x.shape[0]
n = x.shape[1]
```

```python
x = np.vstack((np.ones(n),x))
d = x.shape[0]
n = x.shape[1]
x
```

```python
def derivative(w,x,y):
    return 2*(x @ x.T) @ w - 2 * x @ np.expand_dims(y,axis=1)
```

```python
def gradient_descent(w_old,x,y,learning_rate,iterations):
    w_new = np.zeros((x.shape[0],1))
    for i in range(iterations):
        print(w_old)
        w_new = w_old - (learning_rate / 3) * (derivative(w_old,x,y))
        w_old = w_new
    return w_old
```

```python
w_old = np.zeros((x.shape[0],1))
w = gradient_descent(w_old,x,y,learning_rate=0.001,iterations=100)
```

```python
x_new = np.array([1,9,18])
x_new.T @ w
```
