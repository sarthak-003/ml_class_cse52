```python
data = np.array([
    [1, 2],
    [2, 3.9],
    [3, 5.8],
    [4, 8.1],
    [5, 9.9],
    [6, 12.3],
    [7, 14.1],
    [8, 16.5],
    [9, 18.2],
    [10, 20.4]
]) 
x = data[:, 0].reshape(1,-1)
y = data[:, 1] 

d = x.shape[0]
n = x.shape[1]
```

```python
x = np.vstack((np.ones(n),x))
d = x.shape[0]
n = x.shape[1]
x
```

```python
def updation(w,x,y):
    return 2*(x @ x.T) @ w - 2 * x @ np.expand_dims(y,axis=1)
```

```python
w_updated,w_ini = np.zeros((d,1)) , np.zeros((d,1))

for i in range(1000):
    print(w_ini)
    w_updated = w_ini - (0.001 / 3) * (updation(w_ini,x,y))
    w_ini = w_updated

print(w_updated)
```
