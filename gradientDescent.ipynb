```python
def updation(w,x,y):
    return 2*(x @ x.T) @ w - 2 * x @ np.expand_dims(y,axis=1)
```

```python
w_updated,w_ini = np.zeros((d,1)) , np.zeros((d,1))

for i in range(500):
    print(w_ini)
    w_updated = w_ini - (0.0001 / 3) * (updation(w_ini,x,y))
    w_ini = w_updated

print(w_updated)
```
